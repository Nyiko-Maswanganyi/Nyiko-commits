```python
 import pandas as pd
!wget https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/ml-basics/grades.csv
students = pd.read_csv('grades.csv',delimiter=',',header='infer')

students
students=students.dropna(axis=0,how='any')
students

var_data=students['StudyHours']

def data_distribution(var_data):
    from matplotlib import pyplot as plt
    #two graphs one on top of the other
    fig,ax=plt.subplots(2,1,figsize=(10,4))
    
    #plot a histogram
    ax[0].hist(var_data)
    ax[0].set_ylabel('Frequency')
    
    # Add vertical lines for the mean, median, and mode
    ax[0].axvline(x=var_data.min(), color = 'gray', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=var_data.mean(), color = 'cyan', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=var_data.median(), color = 'red', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=var_data.mode()[0], color = 'yellow', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=var_data.max(), color = 'gray', linestyle='dashed', linewidth = 2)
    
    #plot a box and whiskerswhich will help you evaluate outliers
    ax[1].boxplot(var_data)
    ax[1].set_xlabel('Study Hours')
    
    # Add horizontal lines for the mean, median, and mode
    ax[1].axhline(y=var_data.min(), color = 'gray', linestyle='dashed', linewidth = 2)
    ax[1].axhline(y=var_data.mean(), color = 'cyan', linestyle='dashed', linewidth = 2)
    ax[1].axhline(y=var_data.median(), color = 'red', linestyle='dashed', linewidth = 2)
    ax[1].axhline(y=var_data.mode()[0], color = 'yellow', linestyle='dashed', linewidth = 2)
    ax[1].axhline(y=var_data.max(), color = 'gray', linestyle='dashed', linewidth = 2)
    
    fig.suptitle('Multiple Graphs On One Axis')
    #create another figure
    fig=plt.figure(figsize=(9,4))
     
    #plot a density graph    
    var_data.plot.density()
    plt.xlabel('density distribution')
 
    plt.axvline(x=var_data.mean(), color = 'cyan', linestyle='dashed', linewidth = 2)
    plt.axvline(x=var_data.median(), color = 'red', linestyle='dashed', linewidth = 2)
    plt.axvline(x=var_data.mode()[0], color = 'yellow', linestyle='dashed', linewidth = 2)
    
    #print density distribution graph
    plt.show()
    #print the first two graphs we created
    fig.show()
    
 #outlier at the extreme lower end  
data_distribution(var_data)
#get the 0.01th quartile
q1=students['StudyHours'].quantile(0.01)
#column of hours greater than q1
col=students[students.StudyHours>q1]['StudyHours'] 
             
#plot data using our new function that prints 99.9% of the data,anything else is an outlier and has been removed
data_distribution(col)

#Now lets mess with measures of variance

#use a loop to iterate through our numeric data columns
#print out the range,variance and standard deviation
for bolt in['Grade','StudyHours']:
     print(students[bolt].max()- students[bolt].min())
     print(students[bolt].var())
     print(students[bolt].std())
```
